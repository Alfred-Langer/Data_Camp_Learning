What is Data Visualization?

Creating a graph/chart/diagram to present the insights you've discovered from your data, in an easy to understand manner.

What is Matplotlib?

Matplotlib is a very popular Python package that is used for Data Visualization.

```python
#Import the subpackage pyplot from the package matplotlib
#We use the alias plt for the subpackage pyplot
import matplotlib.pyplot as plt


#List of Data Points for the x-axis
year = [1950,1970,1990,2010]

#List of Data Points for the y-axis
population=[2.519,3.692,5.263,6.972]

#Here we create a line plot using the plot function from pyplot
#We pass in the x-axis and y-axis data points as arguments
plt.plot(year,population)

#After creating the plot we display it using the show function from pyplot
plt.show()
```

```python
#We can add labels to the x-axis and y-axis using the xlabel and ylabel functions
plt.xlabel('Year')
plt.ylabel('Population')

#We can add a title to the plot using the title function
plt.title('World Population Projections')

#Let's remake the plot with the labels and title.
#We can also add a marker to the plot using the marker argument to highilght the data points in the lists
plt.plot(year,population, marker = 'o')

#Let's redipaly the plot with the labels, title and markers
plt.show()
```

```python
#The previous plot was a line plot, we can also create a scatter plot using the scatter function instead of the plot function
plt.scatter(year,population)

#Reapply the labels and title
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('World Population Projections')

#Display the scatter plot
plt.show()


#Notice how we don't need to use the marker argument when creating a scatter plot since the scatter function automatically highlights the data points
```

```python
#Datapoints for the x-axis
number_of_company_employees = [100,1000,10000,100000]

#Datapoints for the y-axis
rating_of_company = [5.0, 6.25, 7.5 , 8.75]


#Create a scatter plot using the scatter function
plt.scatter(number_of_company_employees,rating_of_company)

#Add labels and title
plt.xlabel('Number of Employees')
plt.ylabel('Company Rating')


#You can also have your x-axis or y-axis scale logarithmically rather than linearly using xscale or yscale respectively
#For example, to have the x-axis scale logarithmically
#plt.xscale('log')

#Display the scatter plot
plt.show()

#You can see how the plot would look like if the x-axis was scaled linearly by commenting out the xscale function and running the cell again
```
