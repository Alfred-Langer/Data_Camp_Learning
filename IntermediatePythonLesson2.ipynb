What is a histogram?

A histogram is a type of graph/visualization that is designed to highlight the distribution of your data. In a histogram, the data is divided into intervals known as bins. Each bin represents a range of values, and the height of the bar for each bin indicates the frequency/count of data points that fall within that range. A bar with a small height signifies that fewer data points are in that bin's range, while a taller bar indicates that more data points fall into that bin's range. The pattern of the bar heights provides insight into how the data is distributed across the different bins.



In the histogram above, you have 6 bins with the 1st bin covering the range of scores from 40 - 50, the 2nd bin covering 50 - 60, the 3rd covering 60-70 and so on and so forth. From a basic point of analysis, we can see that a good chunk of the students scored between the range 50 - 70 with 14 scores falling into that range. And the least likely score range was 90 - 100 with only two students achieving a score in that interval.

CREATING A HISTOGRAM USING MATPLOTLIB 

```python
#Import the subpackage pyplot from the package matplotlib
#We use the alias plt for the subpackage pyplot
import matplotlib.pyplot as plt

#Import random module so we can generate random numbers
import random
```

```python
#List containing all of the data points for our dataset
heights_of_college_athletes= []

#Generate 30 random numbers between 60 and 80 and append them to the list
for i in range(30):
    heights_of_college_athletes.append(random.randint(60, 80))
    
#Create a histogram of the data with 5 bins
plt.hist(heights_of_college_athletes, bins=5)

#Add labels to the x-axis and y-axis using the xlabel and ylabel functions and add a title to the plot using the title function
plt.xlabel('Height in Inches')
plt.ylabel('Number of Athletes')
plt.title('Height Distribution of College Athletes')

#Display the histogram
plt.show()
```

```python
'''
There are quite a few options that you can use to customize the appearance of the histogram. 

You can view the documentaion for the hist function by using the help function.
'''

#View the documentation for the hist function
help(plt.hist)

'''You can also view the documentation for the hist function online at https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html'''
```

CUSTOMIZING THE HISTOGRAM

```python
#Create a histogram of the data with 5 bins
#Set the range of the histogram to be between 60 and 80
#Set the color of the histogram to be red
#Set the edgecolor of the histogram to be black
counts, bins, patches = plt.hist(heights_of_college_athletes, bins=5, range=(60, 80), color="red",edgecolor='black')

#plt.hist returns three values, counts, bins, and patches, however these values don't have to be caught in variables if you don't need them
#counts is a list that contains the number of data points in each bin
#bins is a list that contains the edges of the bins
#patches is something I don't quite understand yet but I hope to learn more about it in the future

# Set the y-axis range. We can use the max function to find the maximum value in the counts list and add 2 to it to create consistent padding at the top of the plot
plt.ylim(0, max(counts) + 2)  # Adjust the range as needed

#Add labels to the x-axis and y-axis using the xlabel and ylabel functions and add a title to the plot using the title function
plt.xlabel('Height in Inches')
plt.ylabel('Number of Athletes')
plt.title('Height Distribution of College Athletes')

#Set the x-axis to range from 60 to 80 with ticks every 4 units, the bars of the histogram should now be alligned with the ticks of the x-axis
plt.xticks(range(60, 81, 4))

plt.show()
```
